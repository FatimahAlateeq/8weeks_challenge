### Cleaning Data

```python
# Cleaning 'runners' table:
# Convert 'registration_date' type from object to datetime:
runners['registration_date']=pd.to_datetime(runners['registration_date'])
runners
runners.info()
```

```python
# p1: Cleaning 'customer_orders' table:
# Convert 'order_time' type from object to datetime:
customer_orders['order_time']=pd.to_datetime(customer_orders['order_time'])
# Index 'customer_orders' by adding 'index' column, because of 'exclusions' and 'extras' separating:
customer_orders=customer_orders.reset_index()
customer_orders
customer_orders.info()
customer_orders.isna().sum() # No problem with these nulls for now.
```

```python
# Split function:
def tidy_split(df, column, sep='|', keep=False):
    """
    Split the values of a column and expand so the new DataFrame has one split
    value per row. Filters rows where the column is missing.

    Params
    ------
    df : pandas.DataFrame
        dataframe with the column to split and expand
    column : str
        the column to split and expand
    sep : str
        the string used to split the column's values
    keep : bool
        whether to retain the presplit value as it's own row

    Returns
    -------
    pandas.DataFrame
        Returns a dataframe with the same columns as `df`.
    """
    indexes = list()
    new_values = list()
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        values = presplit.split(sep)
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    return new_df
```

```python
# p2: Cleaning 'customer_orders' table:
# Separating toppings and convert type to int:
extras_sep=customer_orders[['index','extras']]
extras_sep=tidy_split(extras_sep,'extras',sep=', ')
extras_sep['extras']=extras_sep['extras'].astype(int)
extras_sep
exclusions_sep=customer_orders[['index','exclusions']]
exclusions_sep=tidy_split(exclusions_sep,'exclusions',sep=', ')
exclusions_sep['exclusions']=exclusions_sep['exclusions'].astype(int)
exclusions_sep
```

```python
# Cleaning 'runner_orders' table:
# Convert 'pickup_time' type from object to datetime:
runner_orders['pickup_time']=pd.to_datetime(runner_orders['pickup_time'])
# Remove units:
runner_orders['duration'] = pd.to_numeric(runner_orders.duration.str.replace('[^\d.]', ''), errors='coerce')
runner_orders['distance'] = pd.to_numeric(runner_orders.distance.str.replace('[^\d.]', ''), errors='coerce')
runner_orders
runner_orders.info()
runner_orders.isna().sum() # No problem with these nulls for now.
```

```python
# Cleaning 'pizza_recipes' table:
# Separating toppings and convert type to int:
temp = pizza_recipes["toppings"].str.split(', ')
pizza_recipes = pizza_recipes.reindex(pizza_recipes.index.repeat(temp.apply(len)))
pizza_recipes['toppings'] = np.hstack(temp).astype(int)
pizza_recipes
pizza_recipes.info()
```

```python
# Merge some tables:
merg1_df = pd.merge(customer_orders, runner_orders, on='order_id',  how='left')
merg2_df = pd.merge(merg1_df, pizza_names, on='pizza_id',  how='left')
view1=merg2_df
view1.info()
view1
merg3_df=pd.merge(pizza_recipes, pizza_toppings,left_on=['toppings'], right_on = ['topping_id'], how='left')
merg4_df=pd.merge(merg3_df, pizza_names,on=['pizza_id'], how='left')
toppings_table=merg4_df.drop(['toppings'],axis=1)
toppings_table.info()
toppings_table
```
